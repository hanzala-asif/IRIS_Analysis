# Iris Dataset Exploration and Visualization
This notebook demonstrates how to load, inspect, summarize, and visualize the Iris dataset using Python's pandas, matplotlib, and seaborn libraries.

## 1. Setup: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_theme(style="whitegrid")
```

## 2. Load the Dataset
The Iris dataset is loaded from the `Iris.csv` file.

```python
df = pd.read_csv('Iris.csv')
```

## 3. Dataset Structure and Inspection
We will inspect the dataset's shape, column names, and the first few rows to understand its structure.

```python
print("Dataset Shape:", df.shape)
```

```python
print("Dataset Columns:",
      df.columns.tolist())
```

```python
print("First 5 rows:")
df.head(5)
```

## 4. Summary Statistics
Descriptive statistics provide a quick overview of the central tendency, dispersion, and shape of a dataset's distribution.

```python
print("Summary Statistics:")
print(df.describe())
```

## 5. Visualizations


### 5.1 Scatter Plot: Sepal Length vs Sepal Width


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', palette='viridis')
plt.title('Iris Sepal Length vs Sepal Width')
plt.show()
```

### 5.2 Histogram: Distribution of Petal Length
The histogram shows the distribution of petal length, providing insights into how this feature varies across different species.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='PetalLengthCm', kde=True, hue='Species', element="step")
plt.title('Distribution of Petal Length')
plt.show()
```

### 5.3 Box Plot: Petal Width by Species
A box plot is used to display the distribution of petal width for each species, highlighting median, quartiles, and potential outliers.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Species', y='PetalWidthCm', palette='Set2')
plt.title('Box Plot of Petal Width by Species')
plt.show()
```
